```python
choice_widget = widgets.Dropdown(options=["Artist", "Group Year"])
display(choice_widget)

def hist (choice):
    plt.hist(torrents[choice])
choice_widget = widgets.Dropdown(options=["Artist", "Group Year", "Release Type"])
widgets.interact(hist, choice=choice_widget)

def podsumowanie (choice3):
    print(torrents[choice3].describe())
choice_widget3 = widgets.Dropdown(options=["Artist", "Release Type","Group Name", "Total Snatched","Group ID"])
widgets.interact(podsumowanie, choice3=choice_widget3)

def describe_function (choice1):
    print(torrents[choice1].describe())
choice_widget1 = widgets.Dropdown(options=["Artist", "Release Type","Group Name", "Total Snatched","Group ID"])
widgets.interact(describe_function, choice1=choice_widget1)

def box_kind (choice2):
    torrents[choice2].plot(kind = 'box', vert = False, figsize = (14,6))
choice_widget2 = widgets.Dropdown(options=["Group Year", "Total Snatched"])
widgets.interact(box_kind, choice2=choice_widget2)

#przydałoby się uzalężnić bins od ilości danych
def hist (choice3):
    plt.hist(torrents[choice3])
choice_widget3 = widgets.Dropdown(options=["Group Year", "Release Type"])
widgets.interact(hist, choice3=choice_widget3)

def density_function (choice):
    den = torrents[choice].plot(kind = 'density', figsize = (14,6))
    den.axvline(torrents[choice].mean(), color = "red")
    den.axvline(torrents[choice].median(), color = "green")
choice_widget2 = widgets.Dropdown(options=["Group Year", "Total Snatched"])
widgets.interact(density_function, choice=choice_widget2)

torrents["Release Type"].value_counts().plot(kind = "pie", figsize =(6,6))

import keyboard

selected = 1
def show_menu():
    global selected
    print("\n" * 30)
    print("Choose an option:")
    for i in range(1, 5):
        print("{1} {0}. Do something {0} {2}".format(i, ">" if selected == i else " ", "<" if selected == i else " "))
def up():
    global selected
    if selected == 1:
        return
    selected -= 1
    show_menu()
def down():
    global selected
    if selected == 4:
        return
    selected += 1
    show_menu()
show_menu()
keyboard.add_hotkey('up', up)
keyboard.add_hotkey('down', down)
keyboard.wait()
```
